```python
from sklearn import datasets
iris = datasets.load_iris().data

# print(iris)

from sklearn.metrics.pairwise import euclidean_distances
import heapq

def calculateDistance(data1,data2):
    return euclidean_distances(data1,data2)

# calculateDistance(iris)
```

```python
# Agglomerative Implementation

class customAgglomerative:
    
    def __init__(self, data, n_clusters, linkage):
        self.data = data
        self.n_clusters = n_clusters
        self.linkage = linkage
        self.dataset_size = len(self.data)
        self.labels = []
        for i in range (self.dataset_size):
            self.labels.append(-1)
        self.distanceMatrix = calculateDistance(self.data,self.data)
        self.dimension = len(self.data[0])
        self.heap = [] # priority queue
    
    def compute_pairwise_distance(self, data):
        result = []
        dataset_size = len(data)
        for i in range(dataset_size-1):    # ignore last i
            for j in range(i+1, dataset_size):     # ignore duplication
                dist = self.distanceMatrix[i][j]

                # duplicate dist, need to be remove, and there is no difference to use tuple only
                # leave second dist here is to take up a position for tie selection
                result.append( (dist, [dist, [[i], [j]]]) )

        return result
    
    def build_priority_queue(self, distance_list):
        heapq.heapify(distance_list)
        self.heap = distance_list
        return self.heap
        
agglo = customAgglomerative(iris,3,"singleton")
# agglo.compute_pairwise_distance(iris)
agglo.build_priority_queue(agglo.compute_pairwise_distance(iris))   
```
